```python
plt.figure(figsize=(8, 4))
plt.title('Total Count')
plt.ylabel('Runs')
plt.xlabel('Years')
pd.crosstab(df.Date.dt.year,df.Date.dt.year,values=df['Score'],aggfunc='sum').plot(kind='bar',ax=plt.gca())
```

```python
df.Date.dt.year.value_counts().plot(kind='bar')
```

```python
# Bin the data
df['Binned'] = pd.cut(df['Score'], bins=[101, 150,200,250,300], labels=['101-150','151-200','201-250','251-300'])

# Count occurrences of each bin
bin_counts = df['Binned'].value_counts().sort_values(ascending=False)

# Calculate cumulative percentage
bin_counts = bin_counts / bin_counts.sum()
cum_pct = bin_counts.cumsum()

# Plotting
fig, ax1 = plt.subplots()

# Bar plot
ax1.bar(bin_counts.index, bin_counts, align='center', alpha=0.7)
ax1.set_title('Pareto Plot of Runs Scored')
ax1.set_ylabel('Percentage of Runs', color='b')
ax1.set_xlabel('Runs Range')
plt.show()
```
